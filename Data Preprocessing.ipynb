### Shorten the complaint resolution descriptions by using applymap pandas function

```python
nyc_complaints_df['Resolution Description'].unique()
```

```python
def transform_resolution(desc):
    
    '''
    performs mapping in NYC Complaint Dataframe column "Resolution Description" and shortens the comments 
    '''
    
    if(desc == 'The Police Department reviewed your complaint and provided additional information below.'):
        return 'Reviewed'
    elif(desc == 'The Police Department responded to the complaint and took action to fix the condition.'):
        return 'Responded & Took Action'
    elif(desc == 'The Police Department responded to the complaint and with the information available observed no evidence of the violation at that time.'):
        return 'No Evidence'
    elif(desc == "This complaint does not fall under the Police Department's jurisdiction."):
        return 'Does not Fall under Jurisdiction'
    elif(desc == 'The Police Department made an arrest in response to the complaint.'):
        return "Suspect Arrested"
    elif(desc == 'Your report has been submitted. The City will investigate as needed. No further information will be available from 311.'):
        return "Complaint Submitted"
    elif(desc == 'The Department of Sanitation is in the process of investigating this complaint.'):
        return 'Investigation in progress'
        
```

### Filter the data

```python
nyc_complaints_df['Resolution Description'] = nyc_complaints_df['Resolution Description'].apply(transform_resolution)
```

```python
res_type = ['Reviewed', 'Responded & Took Action', 'No Evidence',
       'Does not Fall under Jurisdiction', 'Suspect Arrested',
       'Complaint Submitted', 'Investigation in progress']
```

```python
nyc_complaints_df_new = nyc_complaints_df[nyc_complaints_df['Resolution Description'].isin(res_type)]
```

```python
nyc_complaints_df_new = nyc_complaints_df[['Unique Key', 'Resolution Description']]
```

### Format "created date" column to appropriately visualize in Tableau

```python
def convert_date_to_dayofweek(date):
    d = pd.Timestamp(date)
    return d.day_name()
```

### Calculate Resolution time by taking difference between created date and closed date 

```python
nyc_complaints_df['Created Date'] = pd.to_datetime(nyc_complaints_df['Created Date'])
```

```python
nyc_complaints_df['Closed Date'] = pd.to_datetime(nyc_complaints_df['Closed Date'])
```

```python
nyc_complaints_df.head()
```

```python
nyc_complaints_df['Resolution Time'] = ((nyc_complaints_df['Closed Date'] - nyc_complaints_df['Created Date'])/np.timedelta64(1, 'D'))
```
